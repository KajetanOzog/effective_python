```python
x = np.random.rand(10)
y = np.random.rand(10)

plt.subplot(2,2,1)
plt.scatter(x, y, color=plt.cm.get_cmap('viridis')(np.random.rand(10)))

plt.subplot(2,2,2)
plt.scatter(x,y, s=np.random.uniform(10, 100, 10))

plt.subplot(2,2,3)
plt.scatter(x,y,alpha=np.random.uniform(0,1,10))

plt.subplot(2,2,4)
markers = ['*', 'x', '+', '^', 'v', 's', 'P', '.', ',', '1']
for i, marker in enumerate(markers):
    plt.scatter(x[i], y[i], marker=marker)
plt.show()
```

Poproszę ładny bar plot z lewej, a z prawej pie chart

```python
countries = ['USA', 'Canada', 'Germany', 'UK', 'France']
population = [331, 38, 83, 67, 65]

companies = ['Company A', 'Company B', 'Company C', 'Company D', 'Company E']
market_share = [20, 30, 25, 25, 10]

plt.bar(countries, population, color='green')
plt.title("Population in countries")
plt.xlabel("Country")
plt.ylabel("Population")
plt.show()


plt.pie(market_share, labels=companies, autopct='%1.1f%%')
plt.title("Percent of market share by company")
plt.show()
```

Oraz histogram z zaznaczonym innym kolorem wykresem rozkładu

```python
data = np.random.randn(1000)
plt.hist(data, density=True, bins=20)
x = np.linspace(min(data), max(data), 10_000)
gaussian = (1 / (np.std(data) * np.sqrt(2 * np.pi))) * np.exp(-1 * ((x - np.mean(data)) ** 2) / (2 * np.std(data) ** 2))
plt.plot(x, gaussian, color="black")
plt.show()
```
