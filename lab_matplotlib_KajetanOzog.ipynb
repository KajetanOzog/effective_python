(iiii) każdy punkt oznacz innym markerem

```python
x = np.random.rand(10)
y = np.random.rand(10)

fig, ax = plt.subplots(2, 2, figsize=(15, 8))

ax[0, 0].scatter(x, y, color = plt.cm.viridis(np.random.rand(10)))
ax[0, 1].scatter(x, y, s = np.random.rand(10) * 1000)
ax[1, 0].scatter(x, y, s = 1000, alpha = np.random.rand(10))
markers = ['o', 'x', '+', '*', 's', 'd', 'v', '^', 'p', 'h']
for i, m in enumerate(markers):
    ax[1, 1].scatter(x[i], y[i], marker = m)
plt.show()
```

Poproszę ładny bar plot z lewej, a z prawej pie chart

```python
countries = ['USA', 'Canada', 'Germany', 'UK', 'France']
population = [331, 38, 83, 67, 65]

companies = ['Company A', 'Company B', 'Company C', 'Company D', 'Company E']
market_share = [20, 30, 25, 25, 10]


fig, ax = plt.subplots(1, 2, figsize=(15, 8))
ax[0].bar(countries, population, color = plt.cm.viridis(np.linspace(0, 1, 5)))
ax[0].set_xticks(range(len(countries)))
ax[0].set_xlabel('Countries')
ax[0].set_ylabel('Population')
ax[0].set_xticklabels(countries, rotation=45)

ax[1].pie(market_share, labels=companies, autopct='%1.1f%%', shadow=True, startangle=90)
ax[1].legend()
plt.show()
```

Oraz histogram z zaznaczonym innym kolorem wykresem rozkładu

```python
data = np.random.randn(1000)

plt.hist(data, density=True, bins=30)
x = np.linspace(np.min(data), np.max(data), 10_000)
sigma = np.std(data)
mu = np.mean(data)
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
plt.plot(x, y, color='red')
plt.show()
```

3 wykresy w 3 rzędach. Na pierwszym sin (bardziej grubę) i cos (smooth) [PS albo uzyj from `from scipy.ndimage.filters import gaussian_filter1d` albo `np.convolve` albo `seaborn`]. Na drugim arccos i arcsin^2 kropki i przerywana. Na trzecim scatter danych, ale na dolnym plocie ma być tylko oś X podpisana, a na osi Y wyłącz wartości na osi. Użyj tight layout. Zapisz figurę.

```python
data = np.random.randn(1000)
```

```python
fig, ax = plt.subplots(3, 1, figsize=(15, 8))
ax[0].plot(np.sin(data), label='sin(x)', color='blue', linewidth=5)
ax[0].plot(np.cos(data), label='cos(x)', color='red', linewidth=2)

ax[1].plot(np.arccos((data - np.min(data)) / (np.max(data) - np.min(data))), linestyle="dotted", color='green')
ax[1].plot(np.arcsin((data - np.min(data)) / (np.max(data) - np.min(data))) ** 2, linestyle="dashed", color='purple')

ax[2].scatter(np.random.rand(len(data)), data)
ax[2].get_yaxis().set_visible(False)
plt.show()
```
